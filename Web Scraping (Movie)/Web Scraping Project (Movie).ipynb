```python
all_movies_info = []
all_movies_name=soup.find_all('div',class_='card style_1')

for item in all_movies_name:
    movie_name = item.find('div',class_="content").h2.text
    movie_ratings= item.find('div',class_="user_score_chart")['data-percent']
    movie_link= item.find('div',class_="content").a['href']
    movie_link= "https://www.themoviedb.org"+movie_link
    ur12 = requests.get(movie_link,headers = header)
    mo_url= BeautifulSoup(ur12.text,'html')
    mo_url
    movie_release = mo_url.find('span',class_='release').text.strip().replace('(US)',' ').replace('(IN)',' ').replace('(CZ)',' ').replace('(FR)',' ').replace('(GB)',' ').replace('(KR)',' ').replace('(AU)',' ').replace('(DE)',' ')
    movie_genere = mo_url.find('span',class_='genres').text.strip().replace(',',',')
    movie_runtime = mo_url.find('span',class_='runtime').text.strip()
    director=mo_url.find('li',class_='profile').a.text


my_movies_dict= {
    "Movie Name":movie_name, 
    "Rating":movie_ratings,
    "Relese Date":movie_release,
    "Geners":movie_genere,
    "Runtime":movie_runtime,
    "Director":director,
    "Movies Link ":movie_link
    }
all_movies_info.append(my_movies_dict)
# movies_list.append(my_movies_dict)

print(all_movies_info)
```

```python
mid_facts = []
def select():
    for i in fpage_link:
        header = {'User-Agent':'Mozila/5.0(Windows NT 6.3;Win64;x64)AppleWebkit/537.36(KHTML,like Gecko)Chrome/99.0.4844.51 Safari/537.36'}
        data2 = requests.get(i , headers = header)
        content2 = data2.text
        soup2 = BeautifulSoup(content2 , 'html')
        #print(fsp)
        ffcts = soup2.find_all('section' , class_='header poster')
        for d in ffcts:
            fr = d.find('span' , class_='release').text.replace('\n' , '')
            fgn = d.find('span' , class_='genres').text.replace('\n' , '')
            frt = d.find('span' , class_='runtime').text.replace('\n' , '')
            dr = d.find_all('li' , class_='profile')
            fdr = [dn.get_text() for dn in dr]
            for dtd in fdr:
                if 'Director' in dtd:
                    ddfp = dtd.partition('Director')[0]
                    #print(ddfp)
            data_dct = {
                'Genre':fgn,
                'Release date':fr,
                'Runtime':frt,
                'Director':ddfp.replace('\n' , '')
            }
            #print(fpg_dct)
            mid_facts.append(data_dct)
select()
    
```
