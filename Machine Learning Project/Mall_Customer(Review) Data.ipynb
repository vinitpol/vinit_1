### Creating an Age Group Table

```python
df['Age'].max()
```

```python
bins = np.arange(17,74,4)
```

```python
bins
```

```python
df['Age_Group']=pd.cut(df['Age'],bins)
```

```python
df.head()
```

### Creating an Categories of Segment 

```python
lst = []
m = ''
g = ''
t = ''
ct = ''
sp = ''

for i in df['Spending Score (1-100)']:
    if i > 90:
        sp = 'Spendthrift'
        lst.append(sp)
    elif i > 75:
        ct = 'Centroid'
        lst.append(ct)
    elif i > 50:
        t = 'Target'
        lst.append(t)
        
    elif i > 35:
        g = 'General'
        lst.append(g)
    elif i > 20:
        c = 'Careful'
        lst.append(c)
    elif i >= 1:
        m = 'Miser'
        lst.append(m)
        
        
print(lst)
#lst.append(i)
```

```python
lst
```

```python
segment = []
```

```python
segment = pd.DataFrame(lst)
```

```python
segment['Segment'] = pd.DataFrame(lst)
```

```python
segment
```

```python
segment.drop(columns=0,axis=1)
```

```python
df1 = df.join(segment)
```

```python
df1.drop(columns=0,axis=1,inplace=True)
```

```python
df1['City'].value_counts()
```

### Removing   '.'    from City columns

```python
df1['City'] = df1['City'].replace(['Los Angeles.','Chicago.','Houston.','Phoenix.','Philadelphia.','San Antonio.'],['Los Angeles','Chicago','Houston','Phoenix','Philadelphia','San Antonio'])
```

```python
df1['City'].value_counts()
```

```python
df1.head(7)
```

```python
df1.isnull().sum()
```

### Craeting Generation Column on the bases of Age_Group

```python
gen = []
y = ''
ad = ''
ee= ''
old = ''
for i in df['Age']:
    if i > 17 and i <= 25 :
        y = 'Young'
        gen.append(y)
    elif i > 25 and i <= 35  :
        ad = 'AdultHood'
        gen.append(ad)
    elif i > 35 and i <= 50:
        ee = 'Early Elder'
        gen.append(ee)
    elif i > 50 and i <= 70 :
        old = 'Elder/Old'
        gen.append(old)
    else :
        pass
print (gen)
    
```

```python
df1['Generation'] = pd.DataFrame(gen)
```

```python
df1.isnull().sum()
```

```python
df1.sample(5)
```
